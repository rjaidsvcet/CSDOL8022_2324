```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (3,3)
```

```python
data = pd.read_csv ('./data.csv')
```

```python
data.head()
```

```python
X = data.iloc[:, 0]
y = data.iloc[:, 1]
```

```python
plt.scatter (X, y)
plt.show ()
```

```python
b0, b1 = 1, 1
```

```python
y_pred = b0 + (b1 * X)
```

```python
plt.scatter (X, y)
plt.plot ([min (X), max(X)], [min(y_pred), max(y_pred)], color = 'red')
plt.show ()
```

```python
learningRate = 0.0001
epochs = 1000
```

```python
n = float (len (X))
```

```python
for i in range (epochs):
    del_b1 = (-2/n) * np.sum (X * (y - y_pred))
    del_b0 = (-2/n) * np.sum ((y - y_pred))
    b1 = b1 - learningRate * del_b1
    b0 = b0 - learningRate * del_b0
```

```python
y_pred = b0 + (b1 * X)
```

```python
plt.scatter (X, y)
plt.plot ([min (X), max(X)], [min(y_pred), max(y_pred)], color = 'red')
plt.show ()
```
