```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv ('./tmdb_5000_credits.csv')
movies_df = pd.read_csv ('./tmdb_5000_movies.csv')
```

```python
credits.head ()
```

```python
print (f'Credits : {credits.shape}')
print (f'Movie Dataset : {movies_df.shape}')
```

```python
credits_column_renamed = credits.rename (index=str, columns = {'movie_id' : 'id'})
```

```python
movies_df_merged = movies_df.merge (credits_column_renamed, on='id')
```

```python
movies_df_merged.head ()
```

```python
movies_df_merged.columns
```

```python
movies_cleaned_df = movies_df_merged.drop (columns = ['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
```

```python
movies_cleaned_df.columns
```

```python
movies_cleaned_df.info ()
```

```python
v = movies_cleaned_df.vote_count
m = movies_cleaned_df.vote_count.quantile (0.7)
R = movies_cleaned_df.vote_average
C = movies_cleaned_df.vote_average.mean ()
```

```python
movies_cleaned_df['weighted_average'] = ((R*v) + (C*m))/(v + m)
```

```python
movies_cleaned_df.head ()
```

```python
movies_sorted_ranking = movies_cleaned_df.sort_values ('weighted_average', ascending=False)
```

```python
movies_sorted_ranking.columns
```

```python
movies_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head (10)
```

### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler ()
```

```python
movie_scaled_df = scaler.fit_transform (movies_cleaned_df[['weighted_average', 'popularity']])
```

```python
movie_normalized_df = pd.DataFrame (movie_scaled_df, columns= ['weighted_average', 'popularity'])
```

```python
movie_normalized_df.head ()
```

```python
movies_cleaned_df[['normalized_weighted_average', 'normalized_popularity']] = movie_normalized_df
```

### 50% weightage

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weighted_average'] * 0.5 + movies_cleaned_df.normalized_popularity * 0.5
```

```python
movies_scored_df = movies_cleaned_df.sort_values (['score'], ascending=False)
```

```python
movies_scored_df[['original_title', 'normalized_weighted_average', 'normalized_popularity', 'score']].head (10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfv = TfidfVectorizer (
    min_df = 3,
    max_features = None,
    strip_accents = 'unicode',
    analyzer = 'word',
    token_pattern = r'\w{1,}',
    ngram_range = (1, 3),
    stop_words = 'english'
)
```

```python
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna ('')
```

```python
tfv_matrix = tfv.fit_transform (movies_cleaned_df['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
sig = sigmoid_kernel (tfv_matrix, tfv_matrix)
```

```python
indices = pd.Series (movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_recommendations (title, sig = sig):
    idx = indices[title]
    sig_scores = list (enumerate (sig[idx]))
    sig_scores = sorted (sig_scores, key=lambda x : x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
give_recommendations ('The Dark Knight')
```
